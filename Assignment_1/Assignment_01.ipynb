```python
def getSequenceDictionary(fasta_file_location):
    fastaSeq = open(fasta_file_location).readlines()
    
    seqDict = {}
    seq = ""
    
    for line in fastaSeq:
        if line.startswith(">"):
            seq = line.strip()[1:]
            seqDict[seq] = ""
        else:
            seqDict[seq] += "".join(line.strip().split(" "))
        
    return seqDict
```

```python
def composition(seqDict, perSequence):
    if not perSequence:
        seqComposition = {}
        totalSeq = "".join(list(seqDict.values()))
        
        for i in totalSeq:
            if i in seqComposition:
                seqComposition[i] = (seqComposition[i]*len(totalSeq) + 1)/len(totalSeq)
            else:
                seqComposition[i] = (1/len(totalSeq))
        
        return seqComposition
    else:
        allSeqCompostions = {}
        for seq in seqDict:
            perSequenceCompostion = composition({seq: seqDict[seq]}, False)
            allSeqCompostions[seq] = perSequenceCompostion
        
        return allSeqCompostions
```

### Q5: Overall Compostion

```python
TMBSeqDict = getSequenceDictionary("TMB_output.fasta")
TMHSeqDict = getSequenceDictionary("TMH_output.fasta")
```

```python
TMBOverallComp = composition(TMBSeqDict, False)
TMHOverallComp = composition(TMHSeqDict, False)

print(TMHOverallComp)
print(TMBOverallComp)
```

### Q6: Per Sequence Compostion

#### Q6(a)

```python
TMBPerSeqComp = composition(TMBSeqDict, True)
TMHPerSeqComp = composition(TMHSeqDict, True)

TMBPerSeqComp
```

#### Q6(b) - (f)

```python
truePositives = 0
falseNegatives = 0

for seq in list(TMHPerSeqComp.values()):
    dev_1 = 0
    dev_2 = 0
    for i in TMHOverallComp:
        if i in seq:
            value = seq[i]
        else:
            value = 0
        
        dev_1 += abs(value - TMHOverallComp[i])
        dev_2 += abs(value - TMBOverallComp[i])
    if dev_1 < dev_2:
        truePositives += 1
    else:
        falseNegatives += 1

```

```python
print(truePositives)
print(falseNegatives)
```

```python
trueNegatives = 0
falsePostives = 0

for seq in list(TMBPerSeqComp.values()):  
    dev_1 = 0
    dev_2 = 0
    for i in TMBOverallComp:
        if i in seq:
            value = seq[i]
        else:
            value = 0

        dev_1 += abs(value - TMHOverallComp[i])
        dev_2 += abs(value - TMBOverallComp[i])
    if dev_1 > dev_2:
        trueNegatives += 1
    else:
        falsePostives += 1

```

```python
print(trueNegatives)
print(falsePostives)
```

### Q8: Sensitivity, Specificity and Accuracy

```python
TP = truePositives
FN = falseNegatives
TN = trueNegatives
FP = falsePostives

Sensitivity = TP/(TP+FN)
Specificity = TN/(TN+FP)
Accuracy = (TP+TN)/(TP+TN+FP+FN)

print('Sensitivity using 100% dataset = ', Sensitivity)
print('Specificity using 100% dataset = ', Specificity)
print('Accuracy using 100% dataset = ', Accuracy)
```
